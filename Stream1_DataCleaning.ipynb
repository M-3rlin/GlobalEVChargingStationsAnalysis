```python
def convert_station_id(row):
    station_id = int(row["Charging Station ID"].split("_")[1])
    city = row["Charging Station Location"]
    return f"{city}_{station_id}"


df["station_id"] = df.apply(convert_station_id, axis=1)
df.head()
```

```python
#df.groupby("Charging Station Location")["Chargin Rate (kW)"].avg()

df.groupby(by = ["Charging Station Location"])["Charging Rate (kW)"].sum()
df.groupby(by = ["Charging Station Location","Charger Type"])[["Charging Rate (kW)"]].mean()
```
